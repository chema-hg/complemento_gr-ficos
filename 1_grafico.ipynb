# """ TRAZAR UN SENCILLO GRAFICO DE LINEAS """

```python
# Primero importamos el método pyplot usando el alias plt para no tenerlo que escribir varias veces.
# Luego seguimos otra CONVENCION que es llamar a la función subplots(). Esta función puede generar uno o varios
# Trazados en la misma figura.
# Después usamos el método plot() que intentará trazar los datos que reciba.
# Por último la función plt.show() abre el visor de Matplotlib y muestra el trazado.
```

```python
import matplotlib.pyplot as plt
%matplotlib notebook

class Grafico:
    """Dibuja un gráfico sencillo"""

    def __init__(self):
        self.fig, self.ax = plt.subplots()
        self.squares = [1, 4, 9, 16, 25]

    def run(self):
        self.ax.plot(self.squares)
        plt.show()


if __name__ == "__main__":
    app = Grafico()
    app.run()
```

# """CAMBIAR EL TIPO DE ETIQUETA Y EL GROSOR DE LA LINEA."""

Vamos a usar unas cuantas personalizaciones disponibles para mejorar la legibilidad del trazado.

```python
import matplotlib.pyplot as plt
%matplotlib notebook

class Grafico:
    """Dibuja un gráfico sencillo"""

    def __init__(self):
        self.fig, self.ax = plt.subplots()
        self.squares = [1, 4, 9, 16, 25]

    def run(self):
        self.ax.plot(self.squares, linewidth=3) # (1)
        
        # Establece el título del gráfico y las etiquetas de los ejes.
        self.ax.set_title("Números al Cuadrado", fontsize=24) # (2) 
        self.ax.set_xlabel("Valores", fontsize=14) # (3)
        self.ax.set_ylabel("Cuadrado de los Valores", fontsize=14)
        plt.show()
        
        # Establece el tamaño de las etiquetas de los puntos del eje.
        self.ax.tick_params(axis="both", labelsize=14) # (4)


if __name__ == "__main__":
    app = Grafico()
    app.run()
```

El párametro linewidth de (1) controla el grosor de la línea que genera plot()

El método set_title() de (2) establece un título para el código. Los parámetros sontsize controla el tamaño del texto en los distintos elementos del gŕafico.

Los métodos set_xlabel() y set_ylabel() nos permiten establecer el nombre para cada eje (3) y tick_params() aplica estilo a los puntos. (4) Los argumentos que usamos se aplican a los puntos de los ejes x e y (axis='both') con un tamaño de fuente de 14 (labelsize=14)

El gráfico resultante es más fácil de leer.




# CORREGIR EL TRAZADO

 Aunque ahora el gráfico se ve mejor, los datos no están bien trazados. Si nos fijamos parece que al final del gráfico el valor y de 4 es 25. Esto es porque cuando damos a plot() una secuencia de números, asume que el primer punto de datos corresponde a un valor 0 de la coordenada x, pero nuestro primer punto  corresponde al valor 1 de esa coordenada. Podemos remplazar el comportamiento predeterminado dando a plot() los valores de entrada y salida usados 
para calcular los cuadrados.

```python
import matplotlib.pyplot as plt
%matplotlib notebook

class Grafico:
    """Dibuja un gráfico sencillo"""

    def __init__(self, input_values, values):
        self.fig, self.ax = plt.subplots()
        self.input_values = input_values
        self.squares = values

    def run(self):
        self.ax.plot(self.input_values, self.squares, linewidth=3) # (1)
        
        # Establece el título del gráfico y las etiquetas de los ejes.
        self.ax.set_title("Números al Cuadrado", fontsize=24) # (2) 
        self.ax.set_xlabel("Valores", fontsize=14) # (3)
        self.ax.set_ylabel("Cuadrado de los Valores", fontsize=14)
        plt.show()
        
        # Establece el tamaño de las etiquetas de los puntos del eje.
        self.ax.tick_params(axis="both", labelsize=14) # (4)


if __name__ == "__main__":
    input_values = [1, 2, 3, 4, 5]
    squares = [1, 4, 9, 16, 25]
    app = Grafico(input_values, squares)
    app.run()
```

# Trazar puntos individuales y darles estilo con scatter()

Para dibujar un solo punto utilizamos el método scatter() pasándole como argumentos los valores (x, y) del punto.

```python
import matplotlib.pyplot as plt

plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.scatter(2,4)

plt.show()
```

Si queremos mejorarla un poco podemos añadir un título, etiquetar los ejes y que el texto se lea bien.

```python
import matplotlib.pyplot as plt

plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.scatter(2, 4, s=200) # s es el tamaño del punto

# Establece el título del gráfico y las etiquetas de los ejes.
ax.set_title("Números al cuadrado", fontsize=24)
ax.set_xlabel("Valor", fontsize=14)
ax.set_ylabel("Cuadrados de los Valores", fontsize=14)

# Establece el tamaño de las etiquetas de los puntos de los ejes.
ax.tick_params(axis="both", which="major", labelsize=14)

plt.show()
```

# Trazar una serie de puntos con scatter()

```python
Para trazar una serie de puntos, podemos pasar a scatter() listas separadas de valores x e y de la siguiente forma:
```

```python
import matplotlib.pyplot as plt

valores_x = [1, 2, 3, 4, 5]
valores_y = [1, 4, 9, 16,25]

plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.scatter(valores_x, valores_y, s=100)

# Establece el título del gráfico y las etiquetas de los ejes.
ax.set_title("Números al cuadrado", fontsize=24)
ax.set_xlabel("Valor", fontsize=14)
ax.set_ylabel("Cuadrados de los Valores", fontsize=14)

# Establece el tamaño de las etiquetas de los puntos de los ejes.
ax.tick_params(axis="both", which="major", labelsize=24)
# axis puede coger los valores de "x", "y" o "both"
# el parámetro which puede ser "minor" valor de tamaño menor (que es el estandar en este caso)
# o "major" el que luego especificamos nosotros con la etiqueta labelsize en este caso.

plt.show()
```

## Calcular datos automáticamente

Escribir listas a mano puede ser laborioso, sobre todo cuando tenemos muchos puntos. Es más eficiente usar un bucle.
Vamos a ver como dibujar 1.000 puntos.

```python
import matplotlib.pyplot as plt
%matplotlib notebook

valores_x = range(1, 1001) # secuencia inmutable de números del 1 al 1000
valores_y = [x**2 for x in valores_x] # una compresion de lista [] con todos los números de valores_x al cuadrado.

plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.scatter(valores_x, valores_y, s=5) # s es el tamaño del punto.

# Establece el título del gráfico y las etiquetas de los ejes.
ax.set_title("Números al cuadrado", fontsize=20)
ax.set_xlabel("Valor", fontsize=14)
ax.set_ylabel("Cuadrados de los Valores", fontsize=14)


# Establece el rango para cada eje.
ax.axis([0, 1100, 0, 1_100_000]) # min y max para x y min y max para y

plt.show()

```

## Definir Colores Personalizados

Simplemente añadimos a scatter() el parámetro c="" con el nombre del color dentro o también le podemos pasar un color
RGB (0, 0.8, 0) por ejemplo.

```python
import matplotlib.pyplot as plt
%matplotlib notebook

valores_x = range(1, 1001) # secuencia inmutable de números del 1 al 1000
valores_y = [x**2 for x in valores_x] # una compresion de lista [] con todos los números de valores_x al cuadrado.

plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.scatter(valores_x, valores_y, s=5, c="red") # s es el tamaño del punto.

# Establece el título del gráfico y las etiquetas de los ejes.
ax.set_title("Números al cuadrado", fontsize=20)
ax.set_xlabel("Valor", fontsize=14)
ax.set_ylabel("Cuadrados de los Valores", fontsize=14)


# Establece el rango para cada eje.
ax.axis([0, 1100, 0, 1_100_000]) # min y max para x y min y max para y

plt.show()
```

## Utilizar un mapa de color

Un mapa de color es una serie de colores en un gradiente que van de un color inicial a uno final. Se usan para enfatizar un patrón en los datos. Por ejemplo se podríamos hacer que los valores bajos tengan un color claro y los más altos uno más oscuro. Para usar uno de ellos hay que especificar cómo asignamos a cada punto un color basado en su valor de y.

```python
import matplotlib.pyplot as plt
%matplotlib notebook

valores_x = range(1, 1001) # secuencia inmutable de números del 1 al 1000
valores_y = [x**2 for x in valores_x] # una compresion de lista [] con todos los números de valores_x al cuadrado.

plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.scatter(valores_x, valores_y, s=5, c=valores_y, cmap=plt.cm.Blues) 
# Pasamos la lista de valores y a c y luego indicamos a pyplot que mapa de color debe usar con el argumento
# cmap.


# Establece el título del gráfico y las etiquetas de los ejes.
ax.set_title("Números al cuadrado", fontsize=20)
ax.set_xlabel("Valor", fontsize=14)
ax.set_ylabel("Cuadrados de los Valores", fontsize=14)


# Establece el rango para cada eje.
ax.axis([0, 1100, 0, 1_100_000]) # min y max para x y min y max para y

plt.show()
```

Nota se pueden ver todos los mapas de color disponibles en https://matplotlib.org/stable/gallery/color/colormap_reference.html?highlight=colormap
